```
 #@title <center><b>Movie and TV Shows Template
%mkdir --parents "/content/frames/"
def Posting_Template(imdb, path, folder,screenshot_01,screenshot_02,screenshot_03):
  if folder   !=  "DISABLED":
    media      =  ""
  else:
    media      =  "&media=movie"
  clear_output()
  !rm -f "/root/.nfo"
  !rm -f /content/frames/*
  screenshots = ""
  subprocess.Popen('mediainfo --logfile=/root/.nfo "{}"'.format(path), stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
  if folder    == "DISABLED":
    link       =   path.replace("/content/"+REMOTE_NAME,'')
  if folder    == "Parent Folder":
    link       =   os.path.dirname(path).replace("/content/"+REMOTE_NAME,'')
  if folder    == "GrandParent Folder":
    link       =   os.path.dirname(os.path.dirname(path)).replace("/content/"+REMOTE_NAME,"")
  link         =  !rclone link "{REMOTE_NAME}":"{link}"
  link         =   ''.join(link)
  link         =  "[hidereact=1,2,3,4,5,6,7,8][downcloud]" + link + "[/downcloud][/hidereact]"
  if HIDE_REACT_SCORE != 0:
    link       =  "[hidereactscore={}]{}[/hidereactscore]".format(HIDE_REACT_SCORE,link)
  if HIDE_POST_COUNT != 0:
    link       =  "[hideposts={}]{}[/hideposts]".format(HIDE_POST_COUNT,link)
  thumbnails   =   os.path.dirname(path)
  path         =   os.path.basename(path)
  clear_output()
  thumbnail_01 = ""
  thumbnail_02 = ""
  thumbnail_03 = ""
  if screenshot_01 == True:
    %cd "{thumbnails}"
    clear_output()
    !ffmpeg -hide_banner -loglevel quiet -ss 02:30.0 -i "{path}" -vframes 1 -q:v 2 -y "/content/frames/frame1.png"
    clear_output()
    %cd "/content/frames/"
    clear_output()
    thumbnail_01 =   imgurapi.upload_image("frame1.png")
    thumbnail_01 =  "[img width='800px']" + thumbnail_01.link + "[/img]"
    %cd "/content/"
    clear_output()
  if screenshot_02 == True:
    %cd "{thumbnails}"
    clear_output()
    !ffmpeg -hide_banner -loglevel quiet -ss 05:00.0 -i "{path}" -vframes 1 -q:v 2 -y "/content/frames/frame2.png"
    clear_output()
    %cd "/content/frames/"
    clear_output()
    thumbnail_02 =   imgurapi.upload_image("frame2.png")
    thumbnail_02 =  "[img width='800px']" + thumbnail_02.link + "[/img]"
    %cd "/content/"
    clear_output() 
  if screenshot_03 == True:
    %cd "{thumbnails}"
    clear_output()
    !ffmpeg -hide_banner -loglevel quiet -ss 07:30.0 -i "{path}" -vframes 1 -q:v 2 -y "/content/frames/frame3.png"
    clear_output()
    %cd "/content/frames/"
    clear_output()
    thumbnail_03 =   imgurapi.upload_image("frame3.png")
    thumbnail_03 =  "[img width='800px']" + thumbnail_03.link + "[/img]"
    %cd "/content/"
    clear_output() 
  if screenshot_01 or screenshot_02 or screenshot_03:
    screenshots = "[hr][/hr][indent][size=6][forumcolor][b]Media Screenshots :[/b][/forumcolor][/size][/indent]\n[spoiler='Media Screenshots'][center]{}{}{}[/center]\n[/spoiler]".format(thumbnail_01,thumbnail_02,thumbnail_03)
  if imdb      == "None":
    title      =   path.replace(".", " ").replace("_", " ").replace('(', ' ').replace(')', ' ').replace('}', ' ').replace('{', ' ').replace('[', ' ').replace(']', ' ').replace("'", ' ').replace('@', ' ').replace('  ', ' ').replace('  ', ' ')
    year       =   regex.search(r"(?r) (\d{4}) ",title)[1]
    title      =   title.split(" "+year)[0].lower().split("aka ")[-1].replace(" ","+")
    title      =  !unidecode -c "{title}"
    title      =   title[0]   
    imdb       =   f'http://www.omdbapi.com/?t={title}&y={year}{media}&apikey={API_KEY}&r=json&plot=full'
  else:
    imdb       =   regex.search(r"(tt\d{7,8})",imdb)[1]
    imdb       =   f'http://www.omdbapi.com/?i={imdb}&apikey={API_KEY}&r=json&plot=full'
  with urllib.request.urlopen(imdb) as imdb:
    imdb       =   json.loads(imdb.read().decode())
  title        =   (imdb['Title']+"+"+imdb['Year']).replace("â€“","+").replace(" ","+")
  title        =  !unidecode -c "{title}"
  title        =   title[0]
  yt_link      =   urllib.request.urlopen("https://www.youtube.com/results?search_query="+title+"+Official+Trailer")
  yt_link      =   regex.findall(r"watch\?v=(\S{11})", yt_link.read().decode())
  yt_link      =   yt_link[0]
  text_dump    =  "[center][img]{}[/img]\n".format(imdb['Poster'])
  text_dump   +=  "[forumcolor][b][size=6]{} ({})[/b][/forumcolor]\n".format(imdb['Title'],imdb['Year'])
  text_dump   +=  "[url=https://www.imdb.com/title/{}][img]https://i.imgur.com/FEd1Mx6.png[/img][/url] {}/10 ".format(imdb['imdbID'],imdb['imdbRating'])
  text_dump   +=  "[img]https://i.imgur.com/gvAEcqW.png[/img] {}[/size][/center]".format(imdb['imdbVotes'])
  text_dump   +=  "[hr][/hr][indent][size=6][forumcolor][b]Plot Summary :[/b][/forumcolor][/size]\n\n[justify]{}[/justify][/indent]".format(imdb['Plot'])
  text_dump   +=  "[hr][/hr][indent][size=6][forumcolor][b]Official Trailer :[/b][/forumcolor][/size][/indent]\n\n"
  text_dump   +=  "[center][media=YouTube]{}[/media][/center]".format(yt_link)
  text_dump   +=   screenshots
  text_dump   +=  "[hr][/hr][indent][size=6][forumcolor][b]IMDB Info :[/b][/forumcolor][/size][/indent]\n"
  text_dump   +=  "[list][*][b][forumcolor]Rating :[/b][/forumcolor] {}\n".format(imdb['Rated'])
  text_dump   +=  "[*][forumcolor][b]Genre :[/b][/forumcolor] {}\n".format(imdb['Genre'])
  text_dump   +=  "[*][forumcolor][b]Directed By :[/b][/forumcolor] {}\n".format(imdb['Director'])
  text_dump   +=  "[*][forumcolor][b]Written By :[/b][/forumcolor] {}\n".format(imdb['Writer'])
  text_dump   +=  "[*][forumcolor][b]Starring :[/b][/forumcolor] {}\n".format(imdb['Actors'])
  text_dump   +=  "[*][forumcolor][b]Release Date :[/b][/forumcolor] {}\n".format(imdb['Released'])
  text_dump   +=  "[*][forumcolor][b]Awards :[/b][/forumcolor] {}[/list]".format(imdb['Awards'])
  text_dump   +=  "[hr][/hr][indent][size=6][forumcolor][b]Media Info :[/b][/forumcolor][/size][/indent]"
  !sleep 2s
  with open('/root/.nfo','r',encoding='utf-8') as nfo:
    nfo        =   nfo.readlines()
    del nfo[1]
    nfo[1]     =  "Complete Name                            : {}\n".format(path)
    nfo.pop()
    nfo        =   ''.join(nfo)
  text_dump   +=  "[code title='Media Info']{}[/code]\n".format(nfo)
  text_dump   +=  "[hr][/hr][indent][size=6][forumcolor][b]Download Link :[/b][/forumcolor][/size][/indent]\n[center]{}[/center]".format(link)
  return text_dump
```

```
IMDB          = "None" #@param {type:"string"} 
PATH          = "" #@param {type:"string"} 
FOLDER        = "DISABLED" #@param ["DISABLED","Parent Folder","GrandParent Folder"]
Screenshot_01 = True #@param {type:"boolean"}
Screenshot_02 = True #@param {type:"boolean"}
Screenshot_03 = True #@param {type:"boolean"}
print(Posting_Template(IMDB,PATH,FOLDER,Screenshot_01,Screenshot_02,Screenshot_03))
```
